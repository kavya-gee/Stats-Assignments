```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Loading the Iris dataset using seaborn
iris = sns.load_dataset('iris')

# Displaying the first few rows to get familiar with the data
print(iris.head())

# Randomly sampling 30 observations from the dataset
random_sample = iris.sample(n=30, random_state=42)

# Displaying the randomly sampled data
print(random_sample)

# Repeating random sampling 100 times and calculate the mean sepal length
sample_means = []
for _ in range(100):
    sample = iris.sample(n=30, random_state=np.random.randint(0, 1000))
    sample_means.append(sample['sepal_length'].mean())

print("/n")

# Ploting the distribution of sample means using a histogram
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=10, edgecolor='black')
plt.axvline(x=iris['sepal_length'].mean(), color='red', linestyle='--', label='Population Mean')
plt.title('Distribution of Sample Means for Sepal Length')
plt.xlabel('Mean Sepal Length')
plt.ylabel('Frequency')
plt.legend()
plt.show()

print("/n")

def systematic_sampling(data, sample_fraction):
    population_size = len(data)
    sample_size = int(population_size * sample_fraction)
    step = population_size // sample_size
    indices = np.arange(0, population_size, step)
    return data.iloc[indices]

# Performing systematic sampling (20% of the dataset)
systematic_sample = systematic_sampling(iris, 0.2)

# Ploting original vs sampled data
fig, axes = plt.subplots(1, 2, figsize=(15, 6))


# Systematic sampled data scatter plot
axes[0].scatter(range(len(iris)), iris['sepal_length'], alpha=0.7)
axes[0].set_title('Original Iris Data')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Sepal Length')
axes[0].grid(True)

axes[1].scatter(range(len(systematic_sample)), systematic_sample['sepal_length'], color='red', alpha=0.7)
axes[1].set_title('Systematic Sample of Iris Data')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Sepal Length')
axes[1].grid(True)

plt.tight_layout()
plt.show()

```
