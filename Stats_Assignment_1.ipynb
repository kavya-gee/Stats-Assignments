```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('StudentPerformanceFactors.csv')

print("\nDescriptive Statistics:")
print(df.describe()) # provides an overview of central tendencies and data spread

numerical_columns = df.select_dtypes(include='number').columns

for col in numerical_columns:
    print(f"\nStatistics for {col}:")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    print(f"Standard Deviation: {df[col].std()}")
    print(f"Variance: {df[col].var()}")


# Histogram for Hours Studied Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Hours_Studied'], bins=10, kde=True, color='skyblue')
plt.title('Histogram of Hours Studied')
plt.xlabel('Hours Studied')
plt.ylabel('Frequency')
plt.show()
# As the histogram is skewed right, most students are studying fewer hours

print("/n")

# Box Plot for "Sleep_Hours"
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['Sleep_Hours'], color='lightgreen')
plt.title('Box Plot of Sleep Hours')
plt.ylabel('Sleep Hours')
plt.show()
# Sleep is evenly distributed and no strong impact on scores observed.

print("/n")

# Scatter Plot: Sleep Hours vs Exam Score
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Sleep_Hours'], y=df['Exam_Score'], color='blue')
plt.title('Scatter Plot: Sleep Hours vs Exam Score')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Score')
plt.show()
# There's no strong pattern, sleep may not directly influence scores.

print("/n")

# Scatter Plot: Hours Studied vs Exam Score
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Scatter Plot: Hours Studied vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.axhline(y=df['Exam_Score'].mean(), color='r', linestyle='--', label='Mean Exam Score')
plt.axvline(x=df['Hours_Studied'].mean(), color='g', linestyle='--', label='Mean Hours Studied')
plt.legend()
plt.show()
# Positive correlation is seen, more study hours results in better scores

print("/n")

# Scatter Plot: Attendance vs Exam Score
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Attendance', y='Exam_Score', data=df)
plt.title('Scatter Plot: Attendance vs Exam Score')
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.axhline(y=df['Exam_Score'].mean(), color='r', linestyle='--', label='Mean Exam Score')
plt.axvline(x=df['Attendance'].mean(), color='g', linestyle='--', label='Mean Attendance')
# Strong positive correlation between attendance and scores. Students with low attendance underperform.

# Key Takeaways
# Study Hours Matter: Students who study more tend to score higher in exams.
# Attendance is Key: Higher attendance strongly correlates with better exam performance.
# Sleep Hours: Most students sleep 6â€“8 hours, but sleep has no direct impact on scores.
# Outliers Identified: Students with very low attendance or inconsistent study hours need attention.





```
